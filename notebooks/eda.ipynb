```python
# Remove duplicates
fraud_data.drop_duplicates(inplace=True)
credit_data.drop_duplicates(inplace=True)
ip_country.drop_duplicates(inplace=True)
```

```python
# Convert timestamps to datetime format
fraud_data["signup_time"] = pd.to_datetime(fraud_data["signup_time"])
fraud_data["purchase_time"] = pd.to_datetime(fraud_data["purchase_time"])
```

```python
# Transaction frequency per user
fraud_data["transaction_count"] = fraud_data.groupby("user_id")["user_id"].transform("count")

# Time-based features
fraud_data["hour_of_day"] = fraud_data["purchase_time"].dt.hour
fraud_data["day_of_week"] = fraud_data["purchase_time"].dt.dayofweek
```

```python
# Function to convert IP string to integer
def ip_to_int(ip):
    return int(ip)

# Apply conversion
fraud_data["ip_address"] = fraud_data["ip_address"].astype(float).apply(ip_to_int)
```

```python
# Merge fraud_data with ip_country to get location information
fraud_data = fraud_data.merge(ip_country, left_on="ip_address", right_on="lower_bound_ip_address", how="left")

# Drop unnecessary columns
fraud_data.drop(columns=["lower_bound_ip_address", "upper_bound_ip_address"], inplace=True)

# Check new structure
print(fraud_data.head())
```

```python
# Function to map IP address to country
def map_ip_to_country(ip):
    row = ip_country[(ip_country["lower_bound_ip_address"] <= ip) &
                     (ip_country["upper_bound_ip_address"] >= ip)]
    return row["country"].values[0] if not row.empty else "Unknown"

# Apply the function
fraud_data["country"] = fraud_data["ip_address"].apply(map_ip_to_country)

# Check if country mapping is working
print(fraud_data[["ip_address", "country"]].head(10))
```

```python
# Fraud Distribution
fraud_counts = fraud_data["class"].value_counts(normalize=True) * 100

# Purchase Value Statistics
purchase_value_stats = fraud_data.groupby("class")["purchase_value"].describe()

# Age Statistics
age_stats = fraud_data["age"].describe()

# Hour of Day Distribution
hourly_transactions = fraud_data["hour_of_day"].value_counts().sort_index()

# Day of Week Distribution
daily_transactions = fraud_data["day_of_week"].value_counts().sort_index()

# Create Summary Table
univariate_summary = pd.DataFrame({
    "Feature": ["Fraud Distribution", "Purchase Value (Avg.)", "Age (Avg.)", "Hour of Day (Peak)", "Day of Week (Peak)"],
    "Observation": [
        f"Non-Fraud: {fraud_counts[0]:.2f}%, Fraud: {fraud_counts[1]:.2f}%",
        f"Non-Fraud: {purchase_value_stats.loc[0, 'mean']:.2f}, Fraud: {purchase_value_stats.loc[1, 'mean']:.2f}",
        f"Mean: {age_stats['mean']:.2f}, Min: {age_stats['min']}, Max: {age_stats['max']}",
        f"Peak Hour: {hourly_transactions.idxmax()}, Transactions: {hourly_transactions.max()}",
        f"Peak Day: {daily_transactions.idxmax()}, Transactions: {daily_transactions.max()}"
    ]
})

# Print as table
print(univariate_summary.to_markdown(index=False))
```
